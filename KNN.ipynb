```python
X=df.iloc[:,0:2]
X
```

```python
X=df.iloc[:,0:2].values  #2 dimensional array
X
```

```python
y=df.iloc[:,2] # 1 dimensional series
y
```

```python
y=df.iloc[:,2].values # 1 dimensional  array
y
```

```python
k=5
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=k,metric='euclidean')
```

```python
knn.fit(X,y)
```

```python
pred=[54,160]
```

```python
knn.predict([pred])
```

```python
pred1=[74,160]
```

```python
knn.predict([pred1])
```

```python
y_p=knn.predict(X)
y_p
```

```python
y
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(y,y_p)
```

```python
confusion_matrix(y,y_p)
```

```python
print(classification_report(y,y_p))
```
