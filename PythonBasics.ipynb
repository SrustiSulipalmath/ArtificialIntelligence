```python
marks = int(input("Enter the marks scored : "))
if(marks>100):
  print('invalid marks')
elif(marks==100):
  print('perfect score')
elif(marks>=90):
  print('outstanding')
#if(marks<90 and marks>70):
elif(marks>=70):
  print('Exceeds Excpectations')
elif(marks>=50):
  print('satisfactory')
elif(marks>=0):
  print('poor')
else:
  print('invalid marks')
```
