```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/MLLAB/Datasets/newhousing.csv")
```

```python
print(data.shape)
```

```python
scaler = StandardScaler()
```

```python
scaled_data = scaler.fit_transform(data)
```

```python
pca = PCA(n_components = 2)
reduced_data = pca.fit_transform(scaled_data)
reduced_df = pd.DataFrame(reduced_data,columns=['pc1','pc2'])
```

```python
plt.figure(figsize=(8,6))
plt.scatter(reduced_df['pc1'],reduced_df['pc2'],c='blue',marker='o',edgecolor='k')
plt.title('PCA:DataReduced to 2 Dimensions')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.show()
```

```python
print("\nReduced dimension : ")
print(reduced_df)
print("The reduced dimension is ",reduced_df.shape)
```
