```python
year= [1972,    1982,   1992,    2002,   2012]
e_blr=[100.6,  158.61, 305.54, 394.96, 724.79]
e_mys=[10.5,  25.21,  58.65,  119.27, 274.87]
```

```python
plt.plot(year,e_blr)
plt.plot(year,e_mys)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(year,e_blr,color='red')
plt.plot(year,e_mys,color='black')
plt.show()
```

```python
plt.plot(year,e_blr,color='r')
plt.plot(year,e_mys,color='k')
plt.show()
```

```python
plt.plot(year,e_blr,color='r',label='banglore')
plt.plot(year,e_mys,color='k',label='mysore')
plt.legend()
plt.show()
```

```python
plt.plot(year,e_blr,color='r',label='banglore',linestyle='--')
plt.plot(year,e_mys,color='k',label='mysore',linestyle='-.')
plt.legend()
plt.show()
```

```python
plt.plot(year,e_blr,color='r',label='banglore',linestyle='-')
plt.plot(year,e_mys,color='k',label='mysore',linestyle=':')
plt.legend()
plt.show()
```

```python
plt.plot(year,e_blr,color='r',label='banglore',marker='.',markerfacecolor='k')
plt.plot(year,e_mys,color='k',label='mysore')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot(year,e_blr,color='r',label='banglore',marker='.',markerfacecolor='k')
plt.plot(year,e_mys,color='k',label='mysore')
plt.legend()
plt.xlabel("years")
plt.ylabel("Electricity Consumption")
plt.title("Electricity Consumption Trend")
plt.grid()
plt.show()
```

```python
plt.plot(year,e_blr,color='r',label='banglore',marker='.',markerfacecolor='k')
plt.plot(year,e_mys,color='k',label='mysore')
plt.legend()
plt.xlabel("years",fontsize=13)
plt.ylabel("Electricity Consumption",fontsize=13)
plt.title("Electricity Consumption Trend")
plt.grid()
plt.show()
```

```python
import numpy as np
np.arange(1972,2013,2)
```

```python
plt.figure(figsize=(10,6))
plt.plot(year,e_blr,color='r',label='banglore',marker='.',markerfacecolor='k')
plt.plot(year,e_mys,color='k',label='mysore')
plt.legend()
plt.xlabel("years",fontsize=13)
plt.ylabel("Electricity Consumption",fontsize=13)
plt.title("Electricity Consumption Trend")
plt.xticks(np.arange(1972,2013,2))
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(year,e_blr,color='r',label='banglore',marker='.',markerfacecolor='k')
plt.plot(year,e_mys,color='k',label='mysore')
plt.legend()
plt.xlabel("years",fontsize=13)
plt.ylabel("Electricity Consumption",fontsize=13)
plt.title("Electricity Consumption Trend")
plt.xticks(np.arange(1972,2013,2),rotation=90)
plt.grid()
plt.show()
```
