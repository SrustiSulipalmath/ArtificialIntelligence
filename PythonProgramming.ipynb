```python
def excess_sugar(sugars_g, calories_per_serving):
    # Check if sugar content is more than 10% of total calories
    return sugars_g * 4 > 0.1 * calories_per_serving

def excess_saturated_fat(saturated_fat_g, calories_per_serving):
    # Check if saturated fat is more than 10% of total calories
    return saturated_fat_g * 9 > 0.1 * calories_per_serving

def excess_trans_fat(trans_fat_g, calories_per_serving):
    # Trans fat should ideally be 0, flag if its more than 0
    return trans_fat_g > 0

def excess_sodium(calories_per_serving, sodium_mg):
    # Check if sodium is above 20% of the daily recommended intake (2300 mg for a 2000 calorie diet)
    return sodium_mg > 0.2 * 2300

def excess_calories(food_type, calories_per_serving, serving_size):
    # Define calorie thresholds for excess depending on food type and serving size
    if food_type == "solid":
        return calories_per_serving > 400
    elif food_type == "liquid":
        return calories_per_serving > 200
    return False

# Main function
def get_labels(food_type, serving_size, calories_per_serving, saturated_fat_g, trans_fat_g, sodium_mg, sugars_g):
    # Print messages based on findings
    if excess_sugar(sugars_g, calories_per_serving):
        print("EXCESO AZCARES / EXCESS SUGAR")
    if excess_saturated_fat(saturated_fat_g, calories_per_serving):
        print("EXCESO GRASAS SATURADAS / EXCESS SATURATED FAT")
    if excess_trans_fat(trans_fat_g, calories_per_serving):
        print("EXCESO GRASAS TRANS / EXCESS TRANS FAT")
    if excess_sodium(calories_per_serving, sodium_mg):
        print("EXCESO SODIO / EXCESS SODIUM")
    if excess_calories(food_type, calories_per_serving, serving_size):
        print("EXCESO CALORAS / EXCESS CALORIES")

# Test the function with your input values
get_labels("solid", 40, 150, 0, 0, 150, 16)
```

```python
def percentage_growth(num_users, yrs_ago):
    growth = (num_users[-1] - num_users[-1-yrs_ago])/num_users[-1-yrs_ago]
    return growth

# Do not change: Variable for calculating some test examples
num_users_test = [920344, 1043553, 1204334, 1458996, 1503323, 1593432, 1623463, 1843064, 1930992, 2001078]

# Do not change: Should return .036
print(percentage_growth(num_users_test, 1))

# Do not change: Should return 0.66
print(percentage_growth(num_users_test, 7))
```

```python
def percentage_growth(num_users, yrs_ago):
    growth = (num_users[len(num_users)-1] - num_users[len(num_users)-yrs_ago-1])/num_users[len(num_users)-yrs_ago-1]
    return growth
```
